```python
# Посчитайте процент выживаемости у каждого класса пассажиров (Pclass) - 1 балл
round(df.groupby('Pclass')['Survived'].sum() /
    df.groupby('Pclass')['PassengerId'].count() * 100,
    2
).astype(str) + '%'
```

```python
# Выведите самое популярное мужское и самое популярное женское имя на корабле - 1 балл
men_df = df[df.Sex == 'male']
male_names = men_df['Name'].str.extract('Mr\. ([A-Za-z]+)')[0].dropna()
most_popular_male_name = male_names.value_counts().idxmax()
most_popular_male_name
```

```python
# Выведите самое популярное мужское и самое популярное женское имя на корабле - 1 балл
women_df = df[df.Sex == 'female']
female_names = women_df['Name'].str.extract('\(([^)]+)\)')[0]
fallback_names = women_df['Name'].str.extract('Miss\. ([A-Za-z]+)|Mrs\. ([A-Za-z]+)').bfill(axis=1)[0]
female_names.fillna(fallback_names, inplace=True)
female_names = female_names.str.split().str[0]
most_popular_female_name = female_names.value_counts().idxmax()
most_popular_female_name
```

```python
# Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе - 2 балла
```

```python
men_df = df[df.Sex == 'male']
men_df['MaleName'] = men_df['Name'].str.extract('Mr\. ([A-Za-z]+)')[0]
most_popular_male_names = men_df.dropna(subset=['MaleName']).groupby('Pclass')['MaleName'].agg(lambda x: x.value_counts().idxmax())
most_popular_male_names
```

```python
# Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе - 2 балла

women_df = df[df.Sex == 'female']
female_names = women_df['Name'].str.extract('\(([^)]+)\)')[0]
fallback_names = women_df['Name'].str.extract('Miss\. ([A-Za-z]+)|Mrs\. ([A-Za-z]+)').bfill(axis=1)[0]
female_names.fillna(fallback_names, inplace=True)
female_names = female_names.str.split().str[0]
women_df['FemaleName'] = female_names
women_df = women_df.dropna(subset=['FemaleName'])
most_popular_female_names = women_df.groupby('Pclass')['FemaleName'].agg(lambda x: x.value_counts().idxmax())
most_popular_female_names
```

```python
# Выведите часть таблицы с пассажирами, возраст которых больше 44 лет - 1 балл

df[df.Age > 44]
```

```python
# Выведите часть таблицы с пассажирами, возраст которых меньше 44 лет и которые мужского пола - 2 балла

men_df = df[df.Sex == 'male']
men_df[men_df.Age < 44]
```

```python
# Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека) - 1 балл

df_without_nan_cabin = df.dropna(subset=['Cabin'])
cabin_size_counts = df_without_nan_cabin['Cabin'].apply(lambda x: len(x.split(' '))).value_counts()
cabin_size_counts
```
